#### The strategy

The strategy for this challenge is simple: Get a count of the following:

* The “up” parentheses: `(`
* The “down” parentheses: `)`

The floor that Santa ends up on will be (*number of “up” parentheses*) - (*number of “down” parentheses*).

#### One possible solution: Using a `for` loop to count `(` and `)` characters

One possible solution is to use a `for` to iterate over each character in `raw_input`. Here’s a quick example:

````python
for character in "(()()(()":
    print(character)
````

We can use a loop like the one above to step through each character in `raw_input`, keeping a count of the number of `(` and `)` characters along the way:

````python
open_parentheses = 0
close_parentheses = 0

for character in raw_input:
    if character == "(":
        open_parentheses += 1
    elif character == ")":
        close_parentheses += 1
    else:
        print("Something's wrong with input.")
        print(f"Unexpected character: {character}")
        quit()
        
print(f"Santa arrived on this floor: {open_parentheses - close_parentheses}")
````

I could have written the `if` statement like this...

```
for character in raw_input:
    if character == "(":
        open_parentheses += 1
    else:
        close_parentheses += 1
```

...but I wanted to demonstrate a little “defensive programming” against bad input. In the code above, any character that *wasn’t* a `)` would have increased the “down” count by one. 

The approach I took makes it very clear that only a `(` character would increase the “up” count by one, and only a `)` character would increase the “down” count by one. Any other character is unexpected, and the `else` clause makes it very clear that there’s something wrong with the input.

The `quit()` function causes a Python program to stop immediately.

The result of **280** was correct, and we could proceed to the next part of the challenge.

#### Another possible solution: Using the `count()` method to count `(` and `)` characters

Another solution requires knowing of the existence of the `count()` method, which is built into all strings. Given a *substring* — that’s just fancy talk for a part of a string — `count()` will return the number of occurrences of that substring within the string.

This is one of those moments where *showing* works better than *telling*, so here are some examples:

````python
"aaaaa".count("a")
````

````python
test_word = "abcabcabc"
print(test_word.count("abc"))
````

````python
print(test_word.count("abcabcabc"))
````

Note that `count()` doesn’t count overlapping instances of a substring:

````python
print(test_word.count("abcabc"))
````

With `count()`, getting Santa’s destination floor becomes pretty short:

````python
floor = raw_input.count("(") - raw_input.count(")")
print(f"Santa arrived on this floor: {floor}")
````

### Part two

#### The challenge

> Now, given the same instructions, find the position of the first character that causes him to enter the basement (floor -1). The first character in the instructions has position 1, the second character has position 2, and so on.
> 
> For example:
> 
> - `)` causes him to enter the basement at character position 1.
> - `()())` causes him to enter the basement at character position 5.
>
> What is the position of the character that causes Santa to first enter the basement?

#### The strategy

We used the `for` loop code from part one as the basis for our solution for part two, but with a couple of changes:

* As we go through each character in the input, we need to keep track of what character we’re on. As soon as you hear “keep track of”, you should be thinking “we need a variable for that!”
* We can determine the current floor by calculating `open_parentheses - close_parentheses`. If this value is `-1`, we’re at the basement level! We can print the current position and exit the loop with the `break` statement.

#### The solution

````python
open_parentheses = 0
close_parentheses = 0
current_position = 1

for character in raw_input:
    if character == "(":
        open_parentheses += 1
    elif character == ")":
        close_parentheses += 1
    else:
        print("Something's wrong with input.")
        print(f"Unexpected character: {character}")
        quit()
        
    # See if we’re at the basement level (level -1)
    current_floor = open_parentheses - close_parentheses
    if current_floor == -1:
        print(f"Santa found the basement at position: {current_position}")
        break
        
    current_position += 1
````

#### Building lists with list comprehensions

[**List comprehensions**](https://jakevdp.github.io/WhirlwindTourOfPython/11-list-comprehensions.html) are a way of building new lists from existing ones, using a syntax that’s similar to `for` loops.

Here’s an example that uses a list comprehension to build a list whose values are double those of the original:

#### Using the `pop()` method to retrieve and remove the last item of a list

Here’s an example:

````python
my_numbers = [1, 2, 3, 4, 5]
last_number = my_numbers.pop()
print(f"We popped this number from the end of the list: {last_number}")
print(f"Here’s what the list looks like now: {my_numbers}")
````

#### The strategy

The approach that our initial solution took went like this:

* Pop the last number from the list, and hold onto it.
* Try to add it to every other number in the list, checking to see the two add up to 2020. If you find another such number, stop.
* If, after trying to add that popped number to every other number in the list, you don’t create a pair that sums to 2020, it’s clear that the popped number isn’t one of the two numbers in the list that adds up to 2020. Go back to the first step and try again.

In this process, we reduce the size of the list by one.

Here’s our first solution:

````python
last_number = expenses.pop()
solved = False

while not solved:
    
    for expense in expenses:
        if expense + last_number == 2020:
            print(f"The 2 numbers are {expense} and {last_number}.")
            solved = True
            break

    last_number = expenses.pop()
````

#### The challenge

> The Elves in accounting are thankful for your help; one of them even offers you a starfish coin they had left over from a past vacation. They offer you a second one if you can find three numbers in your expense report that meet the same criteria.
> 
> Using the above example again, the three entries that sum to 2020 are 979, 366, and 675. Multiplying them together produces the answer, 241861950.
> 
> In your expense report, what is the product of the three entries that sum to 2020?

There *is* a way to do this from first principles, but it seems like a lot of work. Isn’t there a better way to do this?

Luckily, this is Python, which comes with a whole lot of libraries, including the [itertools](https://realpython.com/python-itertools/) library, which contains functions for generating *combinations*.

To make the functions in itertools available to your programs, you use Python’s `import` statement. The following imports *all* of itertools’ functions into your programs:

````python
from itertools import *
````

One of the functions that itertools provides is the `combinations()` function, which generates combinations from items in a list.

Here’s a demonstration of `combinations()` that generates all the possible 2-item combinations of numbers in a list.

````python
mu_numbers = [1, 2, 3, 4, 5, 6]
my_combos = list(combinations(my_numbers, 2))
print(my_combos)
````

The combinations are [**tuples**](https://realpython.com/python-lists-tuples/#python-tuples), which are ordered collections of objects. Unlike lists, tuples are *immutable* — you can’t alter their items, or add or remove items from them.

We used `combinations()` to create a list of all the possible combinations of three numbers in the expenses list:

````python
expense_3_item_combos = list(combinations(expenses, 3))
print(expense_3_item_combos)
````

Now that we have these combos, how do we find their sum?

Luckily, Python has a built-in `sum()` function that returns the sum of the numbers in a given tuple or list:

````python
print(sum((2, 3)))
print(sum([7, 8, 9, 10]))
````

We can use `sum()` in a `for` loop to find the combo whose numbers add up to 2020:

````python
for combo in expense_3_item_combos:
    if sum(combo) == 2020:
        print(f"These are the 3 numbers that add to 2020: {combo}")
        break
````

````python
def convert_to_dictionary(line):
    dictionary = {}
    line_as_list = line.split()
    
    min_max = line_as_list[0].split("-")
    dictionary["min"] = int(min_max[0])
    dictionary["max"] = int(min_max[1])
    
    dictionary["char"] = line_as_list[1][0]
    
    dictionary["password"] = line_as_list[2]
    
    return dictionary
````
